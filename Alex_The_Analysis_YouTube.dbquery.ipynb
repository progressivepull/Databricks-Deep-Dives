```sql
-- Count the number of customers for each country and order by customer count in descending order
SELECT `country`, COUNT(`customer_id`) AS `customer_count`
FROM `workspace`.`default`.`customerscsv`
GROUP BY `country`
ORDER BY `customer_count` DESC;


-- Count the total number of products for each category in the productscsv table
SELECT 
    `category`, 
    COUNT(`product_id`) AS total_products 
FROM 
    `workspace`.`default`.`productscsv` 
GROUP BY 
    `category`;

-- Reading Directly from the CSV file by using the name of the  file

SELECT * FROM customerscsv;
```
